它能將一個含任意實數的 K 維向量 z「壓縮」到另一個 K 維實向量 sigma(z) 中，使得每一個元素的範圍都 (0, 1) 之間

![softmax function](https://wikimedia.org/api/rest_v1/media/math/render/svg/e348290cf48ddbb6e9a6ef4e39363568b67c09d3)

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
a = np.array([3.0, 1.0, 0.2])
b = np.array([[1, 2, 3, 6],
              [2, 4, 5, 6],
              [3, 8, 7, 6]])

print(softmax(a))
print(softmax(b))
```

```python
print(softmax(a))
print(softmax(a * 10))
print(softmax(a / 10))
```

```python
# 從 -2.0 ~ 6.0 間，每間隔 0.1 切一段
x = np.arange(-2.0, 6.0, 0.1)

# 然後把變數 x 的值跟定值 1 和 0.2 下去做 softmax
# [x, 1, 0.2] x for -2.0 to 6.0 
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

```python
def plot_softmax(x):
    indexs = np.arange(x.size)
    values = softmax(x)
    labels = x
    
    print(values)
    
    plt.bar(indexs, values)
    plt.xticks(indexs, labels)
    plt.show()
```

```python
plot_softmax(a)
plot_softmax(a * 10)
plot_softmax(a / 10)
```
